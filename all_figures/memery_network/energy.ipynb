```python
import os 
import sys
import matplotlib.pyplot as plt
import numpy as np

def plot_end_to_end(model,dataset,target_acc, path=f"/data/wyz/ForwardFL-Latex/figs"):
    logging_file = logging_file_dict[f"{model} {dataset}"]
    train_time_dict = train_time_dict_dict[model]
    local_data_num = local_data_num_dict[dataset]

    comm_bandwidth = 10 * 1024**2 /8  # k * 1024**2 /8 => k Mb/s 
    energy_to_acc_dict ={}
    
    eval_frequence = 5 if dataset == "yahoo" else 1
    linestyle = [(0, ( )), (0, (3, 1,1,1,1,1)), (0, (4, 1,1,1)), (0, (1, 1)),(0, (2, 4)), (0, (5, 1)),(0,(3,1,3,1)),(0,(4,3,3,1))]*100
    colors=['r','g','b','y','c','m','k',"tan"] * 100
    lw = 3

    def round_to_time(round,eval_frequence,comm_time,method,local_data_num,adhoc=None):
        train_time = train_time_dict[method]
        time1, time2 = 0, 0
        if adhoc == None:
            adhoc = [1,1,1]
        else:
            adhoc = list(map(int,adhoc.split(",")))
        time_list = [0]
        if method == "ours":
            time1 += train_time*local_data_num*adhoc[0]
            time2 += comm_time*local_data_num
            # time = (train_time*local_data_num*adhoc[0] + comm_time*local_data_num)/60
        else:
            time1 += train_time*local_data_num*adhoc[0]
            time2 += 2*comm_time*local_data_num
            # time = (train_time*local_data_num*adhoc[0] + 2*comm_time*local_data_num)/60
        # time_list.append(time)
        for i in range(2,round):
            v_num = adhoc[i-1] if i-1 < len(adhoc) else adhoc[-1]
            if method == "ours":
                time1 += train_time*local_data_num*v_num*eval_frequence
                time2 += comm_time*local_data_num*eval_frequence
                # time += (train_time*local_data_num*v_num*eval_frequence + comm_time*local_data_num)/60
            else:
                time1 += train_time*local_data_num*v_num*eval_frequence
                time2 += 2*comm_time*local_data_num*eval_frequence
                # time += (train_time*local_data_num*v_num*eval_frequence + 2*comm_time*local_data_num)/60
            # time_list.append(time)
        return time1*train_power + time2*comm_power
    def time_to_acc(f,eval_frequence,comm_time,method,local_data_num,target_acc):
        train_time = train_time_dict[method]
        time1, time2 = 0, 0
        max_acc = 0
        acc = [0]
        for line in f:
            if "train_model(): train_model" in line:
                time1 += train_time_dict[method]
            if "'acc'" in line:
                time2 += comm_time * local_data_num * (eval_frequence if len(acc) > 1 else 1)
                max_acc = max(max_acc, float(line[line.rfind("'acc':")+7:line.rfind(",")]))
                acc.append(max_acc)
                if max_acc > target_acc:
                    break
        return acc, time1*train_power + time2*comm_power
        
    def get_time_to_target_acc(target_acc,acc_list,time_list):
        n = len(acc_list)
        i = 0
        while i < n and acc_list[i] < target_acc:
            i += 1
        return time_list[i] if i<n else 10000

    for i,method in enumerate(logging_file):
        file = logging_file[method]
        with open(file) as f:
            max_acc = 0
            for line in f:
                if "Trainable" in line:
                    param_num = int(line[line.rfind(":")+2:-2])
                    print(f"param_num: {param_num}")
                    break
            comm_time = param_num * 4 / comm_bandwidth
            acc = [0]
            if "," in file:
                adhoc = file.split("_")[-1].split(".")[0]
            else:
                adhoc = None
            if method == "ours":
                acc, energy = time_to_acc(f,eval_frequence,comm_time,method,local_data_num,target_acc)
            else:   
                for line in f:
                    if "'acc':" in line:
                        cur_acc = float(line[line.rfind("'acc':")+7:line.rfind(",")])
                        if cur_acc > max_acc:
                            max_acc = cur_acc
                        else:
                            cur_acc = max_acc
                        acc.append(cur_acc)
                        if cur_acc >= target_acc:
                            break
                energy = round_to_time(len(acc),eval_frequence,comm_time,method,local_data_num,adhoc)
            client_num = int(file[file.rfind("client_num")+11:].split("_")[0])
            if acc[-1]<target_acc:
                energy = 10000000
            client_num = 100
            energy_to_acc_dict[method] = [energy,energy*client_num]
    
            # plt.plot(time_list, acc, linestyle=linestyle[i], color=colors[i],label=method,linewidth=lw)
    
    print(f"energy : {energy_to_acc_dict}")
    # per_client_energy
    plt.bar(list(energy_to_acc_dict.keys()), [v[0]/1000 for v in energy_to_acc_dict.values()])

    # 添加标签和标题
    plt.xlabel('Method')
    plt.ylabel('Energy(J)')
    plt.title('Energy cost')
    plt.savefig(f"./figs//eval-energy-per-client-{model}-{dataset}.pdf", bbox_inches="tight")
    plt.show()


    # total
    plt.bar(list(energy_to_acc_dict.keys()), [v[1]/1000 for v in energy_to_acc_dict.values()])
    # 添加标签和标题
    plt.xlabel('Method')
    plt.ylabel('Energy(kJ)')
    plt.title('Energy cost')
    plt.savefig(f"./figs//eval-energy-total-{model}-{dataset}.pdf", bbox_inches="tight")
    plt.show()

    # plt.xlabel("Time",fontsize=20)
    # plt.ylabel("Acc.",fontsize=20)
    # plt.xticks(size = 20)
    # plt.xlim(0,1000)
    # plt.yticks(size = 20)
    # plt.ylim(0,0.98)
    # plt.legend(fontsize=8)
    # plt.title(f"{model} {dataset}",fontsize=30)
    # plt.savefig(f"{path}/eval-e2e-{model}-{dataset}.pdf", bbox_inches="tight")
    # plt.show()

    # # 提取键和值的列表
    # keys = list(time_to_acc_dict.keys())
    # values = list(time_to_acc_dict.values())

    # # 绘制柱状图
    # plt.bar(keys, values)

    # # 设置坐标轴标签和标题
    # plt.xlabel('method')
    # plt.ylabel(f'time to target acc {target_acc}')
    # plt.title('Bar Chart')

    # 显示图形
    # plt.savefig(f"{path}/eval-bar-e2e-{model}-{dataset}.pdf", bbox_inches="tight")
    # plt.show()
    # return time_to_acc_dict
```
