```python
import matplotlib.pyplot as plt
with open("fedFwd_distilbert_agnews_lr0.1_client_num_100_numerical.log","r") as f:
    acc = []
    for l in f.readlines():
        if "quick test accuracy is" in l:
            acc.append(float(l[l.rfind(":")+1:]))
    plt.plot(acc)
```
