```python
import matplotlib.pyplot as plt
file_name = "fedFwd_distilbert_yahoo_lr0.01_client_num_1000_numerical.log"
with open(file_name,"r") as f:
    acc = []
    for l in f.readlines():
        if "quick test accuracy is" in l:
            acc.append(float(l[l.rfind(":")+1:]))
    plt.plot(acc)
plt.title(file_name)
```
