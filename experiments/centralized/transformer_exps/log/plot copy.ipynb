```python
import os 
import sys
import matplotlib.pyplot as plt
import numpy as np

methods = ['new','old']
v_num_list = [100,1000]

for i,method in enumerate(methods):
    for v_num in v_num_list:
        file = f'{method}_method_gpu_agnews_distilbert_forward_lr=0.01_layer=e_v_num={v_num}.log'
        with open(file) as f:
            acc = [0]
            for line in f:
                if "'acc':" in line:
                    acc.append(float(line[line.rfind("'acc':")+7:line.rfind(",")]))
            plt.plot(range(0,len(acc)*10,10), acc, label=f'{method},v_num={v_num}',linewidth=3)
plt.xlabel("step",fontsize=20)
plt.ylabel("Acc.",fontsize=20)
plt.xticks(size = 20)
plt.xlim(0,500)
plt.yticks(size = 20)
plt.ylim(0,0.98)
plt.legend(fontsize=15)
plt.title('DistilBERT+Adapter AGNEWS',fontsize=20)
plt.show()
```

```python
import os 
import sys
import matplotlib.pyplot as plt
import numpy as np

methods = ['new','old']
v_num_list = [1000]

for i,method in enumerate(methods):
    max_acc = 0
    for v_num in v_num_list:
        file = f'{method}_method_gpu_yahoo_distilbert_forward_lr=0.01_layer=e_v_num={v_num}.log'
        with open(file) as f:
            acc = [0]
            for line in f:
                if "'acc':" in line:
                    max_acc = max(max_acc,float(line[line.rfind("'acc':")+7:line.rfind(",")]))
                    acc.append(max_acc)
            plt.plot(range(0,len(acc)*10*5,10*5), acc, label=f'{method},v_num={v_num}',linewidth=3)
plt.xlabel("step",fontsize=20)
plt.ylabel("Acc.",fontsize=20)
plt.xticks(size = 20)
plt.xlim(0,5000)
plt.yticks(size = 20)
plt.ylim(0,0.98)
plt.legend(fontsize=15)
plt.title('DistilBERT+Adapter YAHOO',fontsize=20)
plt.show()
```
